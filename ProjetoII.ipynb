```python
# Gerar o relatório de profiling
csv_files = [
    ('sales.csv', 'Vendas'),
    ('cities.csv', 'Relatório de Profiling - Cities'),
    ('forecast_revenue_real_data.csv', 'Relatório de Profiling - Forecast Revenue Real Data'),
    ('product.csv', 'Relatório de Profiling - Product')
]

for file_path, report_title in csv_files:
    df = pd.read_csv(file_path)

    profile = ProfileReport(df, title=report_title, explorative=True)

    output_file = file_path.replace('.csv', '_profiling_report.html')
    profile.to_file(output_file)

    print(f"Relatório de profiling gerado e salvo como '{output_file}'")
```
